```python
# Data for profitability ratios
data_eng = {
    'Financial Metrics': [
        'Total Assets (Million THB)', 
        'Total Liabilities (Million THB)', 
        'Shareholders Equity (Million THB)', 
        'Issued and Paid-up Capital (Million THB)', 
        'Total Revenue (Million THB)', 
        'Operating Income (Million THB)', 
        'Total Costs and Expenses (Million THB)', 
        'EBIT (Million THB)', 
        'Net Profit (Million THB)', 
        'Earnings per Share (THB)'
    ],
    'YE/63 Dec 31, 2020': [4953.91, 2554.26, 2399.65, 300.00, 3136.08, 3082.51, 2264.81, 871.26, 702.00, 2.34],
    'YE/64 Dec 31, 2021': [4436.40, 1425.81, 2676.37, 300.00, 3363.20, 3273.58, 2325.55, 1099.86, 890.43, 1.48],
    'YE/65 Dec 31, 2022': [4903.03, 1745.28, 2570.29, 300.00, 3726.93, 3589.72, 2614.82, 1113.01, 809.29, 1.35],
    'YE/66 Dec 31, 2023': [5459.40, 2185.33, 2714.49, 300.00, 3756.82, 3684.95, 2652.43, 1109.19, 816.27, 1.36],
    '6M/67 Jun 30, 2024': [6226.75, 2932.01, 2738.37, 300.00, 2068.22, 2018.28, 1485.76, 586.67, 421.70, 0.70]
}

# Creating the DataFrame
df_financials_eng = pd.DataFrame(data_eng)

# Extracting relevant data for plotting
years_eng = ['YE/63 Dec 31, 2020', 'YE/64 Dec 31, 2021', 'YE/65 Dec 31, 2022', 'YE/66 Dec 31, 2023', '6M/67 Jun 30, 2024']
total_assets = df_financials_eng.loc[df_financials_eng['Financial Metrics'] == 'Total Assets (Million THB)', years_eng].values[0]
shareholders_equity = df_financials_eng.loc[df_financials_eng['Financial Metrics'] == 'Shareholders Equity (Million THB)', years_eng].values[0]
net_profit = df_financials_eng.loc[df_financials_eng['Financial Metrics'] == 'Net Profit (Million THB)', years_eng].values[0]
total_revenue = df_financials_eng.loc[df_financials_eng['Financial Metrics'] == 'Total Revenue (Million THB)', years_eng].values[0]

# Calculating ROA
roa = (net_profit / total_assets) * 100

# Calculating ROE
roe = (net_profit / shareholders_equity) * 100

# Calculating Net Profit Margin
net_profit_margin = (net_profit / total_revenue) * 100

# Creating a DataFrame to display the ratios
profitability_ratios = pd.DataFrame({
    'Year': years_eng,
    'ROA (%)': roa,
    'ROE (%)': roe,
    'Net Profit Margin (%)': net_profit_margin
})

# Displaying the DataFrame
print(profitability_ratios)
```

```python
# Reshape data for linear regression model
X = np.array(eps).reshape(-1, 1)  # EPS values as the independent variable
y = np.array(pe_ratio)  # P/E Ratio as the dependent variable

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict P/E Ratio based on the linear regression model
y_pred = model.predict(X)

# Calculate the coefficient of determination (R^2 score) to evaluate the fit
r_squared = model.score(X, y)

# Display the linear regression results
slope = model.coef_[0]  # Slope of the regression line
intercept = model.intercept_  # Intercept of the regression line

slope, intercept, r_squared
```

ขั้นตอนการคำนวณ ROCE:
EBIT (กำไรก่อนหักดอกเบี้ยและภาษี): คำนวณจากกำไรจากการดำเนินงาน
ทุนที่ใช้ในการดำเนินงานทั้งหมด: คำนวณจากส่วนผู้ถือหุ้นรวมกับหนี้สินรวม

```python
# Extracting data for ROCE calculation
ebit = df_financials_eng.loc[df_financials_eng['Financial Metrics'] == 'EBIT (Million THB)', years_eng].values[0]
total_equity = df_financials_eng.loc[df_financials_eng['Financial Metrics'] == 'Shareholders Equity (Million THB)', years_eng].values[0]
total_liabilities = df_financials_eng.loc[df_financials_eng['Financial Metrics'] == 'Total Liabilities (Million THB)', years_eng].values[0]

# Calculating Capital Employed (Total Equity + Total Liabilities)
capital_employed = total_equity + total_liabilities

# Calculating ROCE
roce = (ebit / capital_employed) * 100

# Creating a DataFrame to display the ROCE
roce_df = pd.DataFrame({
    'Year': years_eng,
    'EBIT (Million THB)': ebit,
    'Capital Employed (Million THB)': capital_employed,
    'ROCE (%)': roce
})

# Displaying the ROCE DataFrame
print(roce_df)
```

```python
# Calculating the growth rate of Net Profit for each year

# Function to calculate the growth rate
def calculate_growth_rate(current, previous):
    if previous == 0:
        return 0
    return ((current - previous) / previous) * 100

# Calculating growth rates
growth_rates = []
for i in range(1, len(net_profit)):
    growth_rate = calculate_growth_rate(net_profit[i], net_profit[i - 1])
    growth_rates.append(growth_rate)

# Adding the first year's growth rate as None (since there's no previous year to compare)
growth_rates.insert(0, None)

# Creating a DataFrame to display the growth rates
growth_rate_df = pd.DataFrame({
    'Year': years_eng,
    'Net Profit (Million THB)': net_profit,
    'Growth Rate (%)': growth_rates
})

# Displaying the Growth Rate DataFrame
growth_rate_df
```

```python
# Given data for analysis
ebit_2023 = 1109.19  # EBIT (Earnings Before Interest and Taxes) for 2023 in Million THB
total_equity_2023 = 2500.00  # Assuming Shareholders' Equity in Million THB
total_liabilities_2023 = 2399.82  # Total Liabilities in Million THB for 2023

# Calculating Capital Employed (Total Equity + Total Liabilities)
capital_employed_2023 = total_equity_2023 + total_liabilities_2023

# Calculating ROCE (Return on Capital Employed)
roce_2023 = (ebit_2023 / capital_employed_2023) * 100

# Given data for ROI calculation
net_profit_2023 = 816.27  # Net Profit for 2023 in Million THB
total_investment_2023 = capital_employed_2023  # Assuming total investment equals capital employed

# Calculating ROI (Return on Investment)
roi_2023 = (net_profit_2023 / total_investment_2023) * 100

roce_2023, roi_2023
```

ROCE (Return on Capital Employed): 22.64%

ROI (Return on Investment): 16.66%

บริษัทสามารถสร้างผลตอบแทนได้ 16.66% จากการลงทุนทั้งหมด
การวิเคราะห์:
ROCE ที่ระดับ 22.64% แสดงว่าบริษัทมีความสามารถในการใช้ทุนที่มีอยู่ในการดำเนินงานอย่างมีประสิทธิภาพ ซึ่งหมายความว่าบริษัทสามารถสร้างผลตอบแทนจากทุนที่ใช้ในการดำเนินงานได้ดี

ROI ที่ระดับ 16.66% แสดงว่าบริษัทมีความสามารถในการสร้างผลตอบแทนจากการลงทุนของตน แต่ ROI ต่ำกว่า ROCE ซึ่งอาจบ่งบอกว่าการลงทุนบางส่วนไม่ได้สร้างผลตอบแทนสูงเท่ากับการใช้ทุนในการดำเนินงาน
